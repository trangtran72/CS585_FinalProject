```python
import pandas as pd
import numpy as np
```

```python
def readReview (filename):
    """
    A function to read a movie review file, and return a Pandas data frame
    
    Parameter:
        filename: The txt file contains movie reviews
    Return: A Pandas dataframe
    """
    
    dict ={}
    keys = []
    values = []
    
    f = open(filename, encoding = 'utf-8', errors = 'ignore')
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        row = line.split('\t')
        key = [words for segments in row for words in segments.split(':')][0]
        value =[words for segments in row for words in segments.split(':')][-1]
        keys.append(key)
        values.append(value)
    f.close()
    
    dict = dict.fromkeys(keys, 0)

    index_list = {k:[] for k in keys}
    index_list['product/productId'].append(list(range(0, len(keys), 9)))
    index_list['review/userId'].append(range(1, len(keys), 9))
    index_list['review/score'].append(range(4, len(keys), 9))
    index_list['review/summary'].append(range(6, len(keys), 9))
    index_list['review/text'].append(range(7, len(keys), 9))


    dict['product/productId'] = [values[i] for i in [j for i in index_list['product/productId'] for j in i]]
    dict['review/userId'] = [values[i] for i in [j for i in index_list['review/userId'] for j in i]]
    dict['review/score'] = [values[i] for i in [j for i in index_list['review/score'] for j in i]]
    dict['review/summary'] = [values[i] for i in [j for i in index_list['review/summary'] for j in i]]
    dict['review/text'] = [values[i] for i in [j for i in index_list['review/text'] for j in i]]

    df= pd.DataFrame(dict)
    df = df.dropna(how='all', axis=1)
    
    return df   
```
